```python
import numpy as np; np.random.seed(42)
import matplotlib.pyplot as plt

```

```python
def polar_hist2d(x, y):

    def xy2polar(x,y):
        rr = np.sqrt(x**2 + y**2)
        az = np.arctan2(y,x)
        az[az<0] += 2*np.pi
        return rr, az

    radius, azimut = xy2polar(x,y)

    # define binning
    rbins = np.linspace(0,radius.max(), 30)
    abins = np.linspace(0,2*np.pi, 60)

    #calculate histogram
    hist, _, _ = np.histogram2d(azimut, radius, bins=(abins, rbins))
    A, R = np.meshgrid(abins, rbins)

    # plot
    fig, ax = plt.subplots(subplot_kw=dict(projection="polar"))

    pc = ax.pcolormesh(A, R, hist.T, cmap="magma_r")
    fig.colorbar(pc)
```

```python
azimut = np.random.rand(3000)*2*np.pi
radius = np.random.rayleigh(29, size=3000)
x = radius*np.cos(azimut)
y = radius*np.sin(azimut)

polar_hist2d(x,y)
plt.show()
```
